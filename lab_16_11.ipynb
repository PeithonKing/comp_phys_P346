```python
from library.myrandom import Random
from math import log
import matplotlib.pyplot as plt
```

```python
r = Random(0.2)

half_life = 20
N = 2000
l = log(2)/half_life

t_now = 0
dt = 1
Ns = []
times = []

while t_now<half_life*4:
    prob = l*dt
    for i in range(N):
        fate = r.LCG()
        if fate<prob: N -= 1
    t_now += dt
    Ns.append(N)
    times.append(t_now)

plt.plot(times, Ns)
```

```python
r = Random(0.2)

half_life_a = 20
half_life_b = 40
Na = 10000
Nb = 0
Nc = 0
l_a = log(2)/half_life_a
l_b = log(2)/half_life_b

t_now = 0
dt = 1
Nas = []
Nbs = []
Ncs = []
times = []

while t_now<half_life*10:
    prob_a = l_a*dt
    for i in range(Na):
        fate = r.LCG()
        if fate<prob_a:
            Na -= 1
            Nb += 1

    prob_b = l_b*dt
    for i in range(Nb):
        fate = r.LCG()
        if fate<prob_b:
            Nb -= 1
            Nc += 1
    
    t_now += dt
    Nas.append(Na)
    Nbs.append(Nb)
    Ncs.append(Nc)
    times.append(t_now)

plt.plot(times, Nas)
plt.plot(times, Nbs)
plt.plot(times, Ncs)
```
