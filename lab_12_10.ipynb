```python
filename = "data/fit1.dat"
with open(filename) as f:
    x, y = [], []
    points = f.readlines()
    for point in points:
        s = point.split()
        x.append([float(s[0])])
        y.append([float(s[1])])
x = Matrix(x, name = "x")
y = Matrix(y, name = "y")
```

```python
a = polynomial_fit(x, y, 3)
print(a)
```

```python
coeff = a.T().mat[0][::-1]
plot_x = m.arange(min(x.T().mat[0]), max(x.T().mat[0]), 0.1)
plot_y = [P(i, coeff) for i in plot_x]
plt.scatter(x.mat, y.mat)
plt.plot(plot_x, plot_y)
```
