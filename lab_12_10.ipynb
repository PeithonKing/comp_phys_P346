```python
import matplotlib.pyplot as plt
from library.matrix import Matrix
import library.matrix as m
from library.nonlinear_equations import P
from library.fitting import fit_data
```

```python
filename = "data/fit2.dat"
with open(filename) as f:
    x, y = [], []
    points = f.readlines()
    for point in points:
        xs, ys = map(float, point.split())
        x.append([xs])
        y.append([ys])
    x = Matrix(x, "x")
    y = Matrix(y, "y")


# saving as a csv file
# import pandas as pd
# data = list(zip(x.T().mat[0], y.T().mat[0]))
# datap = pd.DataFrame(data, columns=["x", "y"])
# print(datap)
# datap.to_csv(f"{filename[:-4].csv}", index=False)
```

```python
a = fit_data(x, y, 3)
print(a)
```

```python
coeff = a.T().mat[0][::-1]
plot_x = m.arange(min(x.T().mat[0]), max(x.T().mat[0]), 0.1)
plot_y = [P(i, coeff) for i in plot_x]
plt.scatter(x.mat, y.mat)
plt.plot(plot_x, plot_y)
```
