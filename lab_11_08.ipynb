# Determining value of pi

```python
# With Illustration
seed = 0.69
r = Random(seed = seed)
inside = []
outside = []
n = 10**5
for i in tqdm(range(n)):
    r1 = r.LCG()
    r2 = r.LCG()
    if (r1-0.5)**2 + (r2-0.5)**2 <= 2**-2:  # circle with radius = 1/3 and center at (0.5, 0.5)
        inside.append([r1, r2])
    else:
        outside.append([r1, r2])
print(len(inside)*4/n)
```

```python
inside_x, inside_y = zip(*inside)
outside_x ,outside_y = zip(*outside)
plt.figure(figsize = (6,6))
plt.plot(outside_x, outside_y, '.')
plt.plot(inside_x, inside_y, ".")
plt.show()
```
