# Determining value of pi

```python
# With Illustration
seed = 0.1
inside = []
outside = []
n = 10**7
for i in tqdm(range(n)):
    r1 = LCG(seed)
    seed = r1
    r2 = LCG(seed)
    seed = r2
    if (r1-0.5)**2 + (r2-0.5)**2 <= 3**-2:  # circle with radius = 1/3 and center at (0.5, 0.5)
        inside.append([r1, r2])
    else:
        outside.append([r1, r2])
print(len(inside)*9/n)
```

```python
inside_x, inside_y = zip(*inside)
outside_x ,outside_y = zip(*outside)
plt.plot(outside_x, outside_y, 'bo')
plt.plot(inside_x, inside_y, 'ro')
plt.show()
```
